```python
%sql

CREATE CATALOG if not EXISTS weather;
use catalog weather;
```

```python
%sql 
create schema if not exists weather.gold;
create schema if not exists weather.silver;
create schema if not exists weather.bronze;
```

```python
%sql
select count(*) from weather.bronze.mussoorie_1980_2026;
```

```python
from pyspark.sql.functions import row_number, year
from pyspark.sql.window import Window

df = spark.table("weather.bronze.mussoorie_1980_2026")
df_with_year = df.withColumn("year", year(df["time"]))
window_spec = Window.partitionBy("year").orderBy("time")

df_limited = df_with_year.withColumn("rn", row_number().over(window_spec)).filter("rn <= 10").drop("rn")
display(df_limited)
```
